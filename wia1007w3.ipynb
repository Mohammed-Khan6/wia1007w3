```python
# cell 1: import
import numpy as np
np.__version__  # shows numpy version
```

```python
# 2x2 array (example with random floats)
arr_2x2 = np.random.random((2,2))
print("2x2 array:\n", arr_2x2)
```

```python
arr_3x4 = np.random.random((3,4))
print("3x4 array:\n", arr_3x4)
```

```python
ones_3x3 = np.ones((3,3))
zeros_2x4 = np.zeros((2,4))
print("ones_3x3:\n", ones_3x3)
print("zeros_2x4:\n", zeros_2x4)
```

```python
# choose dimensions
n = 4
p = 5
m = 3

A = np.random.random((n,p))
B = np.random.random((p,m))

# matrix multiplication: result shape (n,m)
C = np.dot(A, B)     # or A @ B
print("A shape:", A.shape)
print("B shape:", B.shape)
print("A dot B shape:", C.shape)
print("A dot B:\n", C)
```
